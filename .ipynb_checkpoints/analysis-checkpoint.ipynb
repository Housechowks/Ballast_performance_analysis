# Cale-Ballast Contract Perfomance Analysis for the Month of August and September
### This analysis aims to evaluate progress of the said project and evaluate its feasibility.
#### focusing on 3rd party trucks and thier owners performance profile.

```python
import pandas as pd
import os
```

###  Task 1:
#### Since deliveries are done per  24hrs shifts,  merging individual csvs to one master csv sheet takes precedence here.

```python
df = pd.read_csv("./deliveries/020821-deliveries.csv")

files = [file for file in os.listdir('./deliveries')]

all_deliveries = pd.DataFrame()

for file in files:
    df = pd.read_csv("./deliveries/"+file)
    all_deliveries = pd.concat([all_deliveries,df])
    
all_deliveries.to_csv("all_deliveres.csv", index=False)

all_deliveries.head()
```

### Task 2:
#### data clean up 

```python
all_deliveries =all_deliveries.drop('Unnamed: 0',1)
all_deliveries =all_deliveries.drop('Unnamed: 6',1)
all_deliveries.head()
```

### Task 3:
#### augmenting data for analysis purposes

```python
# setting this columns to numeric values

all_deliveries['Del. No'] = pd.to_numeric(all_deliveries['Del. No'])
all_deliveries['Tonagge'] = pd.to_numeric(all_deliveries['Tonagge'])


# source fro the specific month

all_deliveries['month']= all_deliveries['Date'].str[0:1]
# source for the specific day

all_deliveries['day']= all_deliveries['Date'].str[2:4]


#finding the total amount per delivery which is Kenya shillings (Ksh)

all_deliveries['total_amount'] = all_deliveries['Tonagge']* 170

# sorting deliveries per truck

all_deliveries =all_deliveries.sort_values(['Veh. Reg'])

#all_deliveries.drop(columns = 'day_month', inplace=True)


all_deliveries.head()
```

## Data plots showing relationships

### Task 4:
#### Which specific material had the highest demand, and which generated the most revenue?

```python
all_deliveries =all_deliveries.sort_values(['Specifications'])
all_deliveries['day'] = pd.to_numeric(all_deliveries['day'], errors='coerce')


all_deliveries.head(500)
all_deliveries
```

```python
results = all_deliveries.groupby('Specifications').sum()

all_deliveries['day'] = all_deliveries['day'].fillna(0)

results.head(30)
```

```python
import matplotlib.pyplot as plt

specific_material_delivered =  [material for material, df in  all_deliveries.groupby('Specifications')] 

plt.bar(specific_material_delivered, results['total_amount'] , color='yellow',edgecolor='black')
plt.xticks(specific_material_delivered, rotation='vertical', size=8, color='blue')
plt.ylabel('total amount in (ksh)')
plt.xlabel('specific material supplied')
plt.show()
```


#### > a plot of total amount generated by each specific material over the stipulated time.
#### > Its pretty much evident that 0-40 had the highest demand and generated the highest revenue keeping production cost constant.


### Task 5:
#### when was suppy at its peak?

```python
days_results = all_deliveries.groupby('day').sum()
days_results.head()
```

```python
each_day =  [day for day, df in  all_deliveries.groupby('day')] 

plt.plot(each_day, days_results['total_amount'] , color='yellow')
plt.xticks(each_day, size=8, color='blue')
plt.ylabel('Total Amount (ksh. *10^6)')
plt.xlabel('Days which supply was made')
plt.grid()
plt.show()
```

#### > for purposes applications and data coleration ignoring data without datestamp was propoesd and inclusion recomended in overall framwork. 

#### supply was at its peak around mid month 

### Task 6:
### which trucks have the best performance and consistence?

```python
all_deliveries.dropna(how='any')
truck_results= all_deliveries.groupby('Veh. Reg').sum()
truck_results = truck_results[[ 'Tonagge','total_amount']]
truck_results
```

```python
truck =  [truck for truck, df in  all_deliveries.groupby('Veh. Reg')] 

plt.plot(truck, truck_results['total_amount'], color='g')
plt.xticks(truck, size=9, rotation='vertical', color='blue')
plt.ylabel('Total Amount (ksh)')
plt.xlabel('trucks')
plt.rcParams["figure.figsize"]=(19,5)
plt.grid(color= "black")
plt.show()
truck_results.loc[truck_results['total_amount'].idxmax()]
```
